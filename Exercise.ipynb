# Data Science Process
1. Ask a question
	- What is the goal?
	- What woul dyou do if you had all data?
	- What do you want to predice or estimate?

Think of the relationship between different variables; how does x impact y?
	- Does eating dinner late impact your ability to fall asleep early?
	- Do the most popular songs in the US affect what is most popular song worldwide?

2. Determine the necessary data
	- What sample size is needed?
	- What data do we need to prove or refute the hypothesis

When forming your hypothesis, it's important to find data that will disprove your hypothesis.  It's infeasible to to be 100% correct with, so what you need to do is be highly certain event x will impact event y.
	- Margin of error: The maoun that the results of our survey will differ from the real population value
	- Confidence level: The probability that if we were to run another survey with the same metrics it would return the same results
	- Population size: Size of the population we're collecting data on
	- Likely sample proportion: The percentage of people surveyed whose results we anticipate matching the expected outcome

Our Date-a-Scientist dataset has 200 entries, meaning that our current sample size is equal to 200. If we have a 5% margin of error, 95% confidence level, and a likely sample proportion of 50%, what size does our overall population need to be?
	A. 132
What if the population of Date-a-Scientist users is closer to 5,000? Keeping the same values for margin of error, confidence level, and likely sample proportion, what should the size of our sample be?
	A. 367
This exercise included a calculator that determined what hte sample size should be.  It would be nice to have the algorithms available.

3. Get the data
	- How were the data sampled?
	- Which data are relevant
	- Are there privacy issues?

4. Clean and organize data
	- Is the data readable
	- Get rid of uncessary values
	- Transform tabular data

5. Explore data
	- Plot the data
	- Are there anomalies?
	- Are there patterns?

6. Model data
	- Build a model
	- Fit the model
	- Validate the model

7. Communicate your findings
	- What did we learn?
	- Do the results make sense?
	- Can we tell a story?
